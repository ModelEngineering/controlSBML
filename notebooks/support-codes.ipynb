# Exponential Filter

```python
def filterSignal(u_vals, kF):
    """
    Filters signal by the parameter value.
    
    Parameters
    ----------
    u_vals: list-float
    kF: float in [0, 1]
    
    Returns
    -------
    list-float    
    """
    y_vals = [u_vals[0]]
    for idx, val in enumerate(u_vals[1:]):
        y_vals.append(kF*u_vals[idx] + (1 - kF)* y_vals[idx-1])
    return np.array(y_vals)

# Tests
size = 100
arr = np.array(range(size))
u_vals = size*(np.random.rand(size) - 0.5) + arr
x_vals = range(size)
plt.plot(x_vals, filterSignal(u_vals, 0.5), c="red")
plt.ylim([0, size])
plt.plot(x_vals, arr, linestyle="--", c="black")
```

```python
def evalFilter(kF, slope=1, size=100, ax=None):
    if ax is None:
        _, ax = plt.subplots(1)
    x_vals = range(size)
    # True signal
    arr = slope*np.array(range(size))
    ax.plot(x_vals, arr, linestyle="--", c="black")
    # True + noise
    u_vals = size*(np.random.rand(size) - 0.5) + arr
    ax.plot(x_vals, u_vals, c="blue")
    # True + filtered-noise
    ax.plot(x_vals, filterSignal(u_vals, kF), c="red")
    #
    ax.set_xlabel("time")
    ax.set_ylabel("value")
    ax.legend(["True", "True+Noise", "True+Filtered-noise"])
    ax.set_title("slope=%2.2f, kF=%2.3f" % (slope, kF))
    
evalFilter(0.9, slope=0.5, size=100)
```

```python
_, axes = plt.subplots(2, 2, figsize=(10,10))
for idx1, slope in enumerate([0.5, 2.0]):
    for idx2, kF in enumerate([0.9, 0.1]):
        evalFilter(kF, slope=slope, ax=axes[idx1, idx2])
```
