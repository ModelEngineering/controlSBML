# Nyquist

```python
tf = control.TransferFunction([1], [1, -5])
tf
```

```python
cl_tf = control.feedback(tf, 1)
```

```python
sys_tf_ = 1/(s - 5)
cl_tf_ = sympy.simplify(sys_tf_/(1 + sys_tf_))
cl_tf_
```

So, the closed loop system with loop transfer function $\frac{1}{s - 5}$ has the transfer function $\frac{1}{s-1}$, which has a closed loop pole at 4 and so is unstable. But the Nyquist plot does not show this.

```python
control.nyquist_plot([tf])
```

Interpretation
1. There is an open loop pole on the positive real axis.
2. $N$, the number of clockwise encirclements of -1, $N = Z - P$, where $P$ is the number of open loop poles.
3. So, $0 = Z - 1$ or $Z = 1$. There is one open loop zero at -1 and so there is an unstable closed loop pole.
